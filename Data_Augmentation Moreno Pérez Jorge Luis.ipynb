# Data Increasing
A common problem, is the quantity of data that is available to find or enter to the models,to obtain the results that we want it is necessary that the trainment happens with a extremely huge quantity of data (dataset), but in a lot of cases it is not possible to be able to have the required amount of data to complete the tasks.
Here is when **Data Augmentation** technique appears, and as its name says, it gives us the chance to increase our date in two different ways:




*   Adding changes on the original data. For example: Taking a centered original picture, we duplicate it off-center, changing axles...
*   Using different distributions. For example: If we want to train a model to classify high resolution images, we add low resolution images with "noise"... always keeping the best proportion of high quality.

The Keras library gives the ability to use the data Augmentation automatically when training a model.
This is achieved thanks to the use of the ImageDataGenerator class.
There are admitted many variety of techniques,as well as pixel scales, such as:



*   The changes on the image (translations) through the arguments  **width_shift_range** and **height_shift_range**.
*  Flip the image through the arguments **horizontal_flip** and **vertical_flip**.
*   Images rotations through the argument **rotation_range**
*   Image bright through the argument **brightness_range**.
*   Image zoom through the argument **zoom_range**.

Let's see in an independently way examples of images after they were increased. It is common to use a variety of techniques at the same time during the training.


```
import matplotlib.pyplot as plt    #Library that allows graph, such as MATLAB.
import numpy as np                  #Library that allows work with arrangement.  
from numpy import expand_dims       #Expands the arrangement way.
from keras.preprocessing.image import load_img    #Of Kera´s library to Data Augmentatios, we upload the image on the python library images format(PIL).
from keras.preprocessing.image import img_to_array #It will take an image(PIL) and transform it in a matriz. If it starts from a PNG image, the values insede the image will be between 0 and 255.
from keras.preprocessing.image import ImageDataGenerator #class that takes original data, it transforms them in aleatory way and gives back just the new transformaded data.
from PIL import Image #Python Images Library
from urllib.request import urlopen #It allows to read an electronic direction
```

First, we upload an image from internet


```
url = 'http://tuidoloesunforro.com.ar/wp-content/uploads/2017/11/slash.jpg'
filename = urlopen(url)
img = Image.open(filename)
plt.imshow(img)
```

Let's start with an image change that means move all the image pixels in a direction, like horizontal or vertically, while the same dimensions of the image are kept.

The arguments width_shift_range and height_shift_range of the ImageDataGenerator builder control the quantity of horizontal and vertical displacement, respectively.


```
# We convert the image to an arrangement
data = img_to_array(img)
# We make an expansion to a sample 
samples = expand_dims(data, 0)
# We create the Data Increaser just using the image change 
datagen = ImageDataGenerator(width_shift_range=[-200,200])
# We prepare the iterator
it = datagen.flow(samples, batch_size=1)
# We generate smaples and we show the results
for i in range(9):
	plt.subplot(330 + 1 + i)
	# Lots of images are generated
	batch = it.next()
	image = batch[0].astype('uint8') #It becomes a valid value to show it as an image 
	plt.imshow(image)
	plt.tight_layout()
plt.show()
```

A flip means to reverse the rows or columns of pixels in the case of an horizontal or vertical flip, respectively.
The increase is specified through an boolean argument **horizontal_flip** or **vertical_flip** for the class **ImageDataGenerator**.

```
datagen = ImageDataGenerator(horizontal_flip=True)
it = datagen.flow(samples, batch_size=1)
for i in range(9):
	plt.subplot(330 + 1 + i)
	batch = it.next()
	image = batch[0].astype('uint8')
	plt.imshow(image)
	plt.tight_layout()
plt.show()
```

The quantity of zoom is sampled evenly aleatically of the zoom region from each dimension (width, height) separated.

Float or [inferior limit, superior limit]. Range from aleatory zoom. If it is a float, [inferior limit, superior limit] = [1-zoom_range, 1 + zoom_range]

```
datagen = ImageDataGenerator(zoom_range=.5) #float or range [lower,upper]
it = datagen.flow(samples, batch_size=1)
for i in range(9):
	plt.subplot(330 + 1 + i)
	batch = it.next()
	image = batch[0].astype('uint8')
	plt.imshow(image)
	plt.tight_layout()
 
plt.show()
```

The image is turned randomly in the sense of clockwise with a determined number or degrees from 0 to 360. It is possible that the pixels flip outside the image frame and leave frame areas without pixel data that must be completed.

```
datagen = ImageDataGenerator(rotation_range=220)
it = datagen.flow(samples, batch_size=1)
for i in range(9):
	plt.subplot(330 + 1 + i)
	batch = it.next()
	image = batch[0].astype('uint8')
	plt.imshow(image)
plt.show()
```

# TextAttack
Is a framework of Python from adversary attack, adversary training and Data Augmentation in NLP. We will just focus on the Data Augmentation.

```
!pip install textattack
```

```
from textattack.augmentation import WordNetAugmenter, EmbeddingAugmenter, EasyDataAugmenter, CharSwapAugmenter
```

```
text = "Easy peasy lemon squezzy"
```

TextAttack gives us four methods for the Data Augmentation.


## WordNetAugmenter
Increase the text replacing it with the synonym of the thesaurus (synonyms dictionary) of WordNet.

WordNet is a lexical database of semantic relations between words in more than 200 languages. The sustantives, verbs, adjectives and adverbs get together in sets of cognitive synonyms (synsets), each one expresses a different concept.
WordNet seems superficially as a thesaurus, because it groups the words based on the meaning.



```
aug = WordNetAugmenter();
aug.augment(text)
```


## EmbeddingAugmenter
 Increases the text transforming words with scale, with a restriction to guarantee that its cosine similarity be at least 0.8.


```
aug = EmbeddingAugmenter();
aug.augment(text)
```

## CharSwapAugmenter
Increases the text substituting, eliminating, inserting and exchanging adjacent characters.

```
aug = CharSwapAugmenter()
aug.augment(text)
```


## EasyDataAugmenter
This increases the text with a mixture of replacement of Wordent synonymous, eliminating words, changing the order of them and insertion of synonymous. All of these 4 functionalities happen randomly. So, we will obtain different results every time when we run the code. This brings back 4 increased results, in comparison with the other methods.

```
aug = EasyDataAugmenter()
aug.augment(text)
```

Googletrans uses the Google Translate Ajax API to the detections and translation of languages.

The principal parameters for the method **translate ()** are:



1.   **src**: Source language. This parameter is optional becauae **googletrans** will detect it.
2.   **dest**: Destiny language. Obligatory parameter.

1.   **text**: The text that is going to be translated from the Source Language to the Destiny Language. Obligatory parameter.





 



When using this method what happens is a retranslation, we pass from the original language to a destiny language and then, we go back to the original. During this retranslation, there is a slight change on the phrase between the original text and the retranslated one but the meaning is the same.

The supported languages can be consulted in:https://cloud.google.com/translate/docs/languages?hl=es

To obtain the best results it is recommended to do some tests with the languages, if the destiny language is too similar the translation does not lose information, however it might be a problem with the translation being this one too literal, this becomes a little bit a situation of try and failure.
